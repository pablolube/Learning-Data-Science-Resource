## Intro a series de Tiempo

```python
import pandas as pd
#convertir fechas que estaban con otro formato (ejemplo tipo string) a tipo datetime : pd.to_datetime

fecha_inicial='03/01/2020'
print(fecha_inicial,type(fecha_inicial))

fecha_modificada = pd.to_datetime(fecha_inicial,dayfirst=True) #parametro dayfirst=True significa que el formato de la fecha que le pasamos es dia/mes/año
print(fecha_modificada,type(fecha_modificada))
```

**CREA POR RANGO DE FECHAS**

Si le pones solo fecha de inicio y fin por default te lo toma diario.
Podemos en vez de ponerle fecha de fin una variable que sea perirodos

```python
fecha = pd.to_datetime('03/01/2020',dayfirst=True) 
fecha
```

```python
mes_inicio = fecha.strftime('%Y-%m')
mes_inicio
```

```python
# si queremos crear unas fechas dentro de un rango: pd.date_range

fin = pd.to_datetime('10/10/2020',dayfirst=True) # fecha final es 2020-10-10 00:00:00

fechas_1 = pd.date_range(start=fecha_modificada, end=fin, freq='D') #me crea un rango DIARIO(freq='D') desde fecha inicio (start) hasta fecha final (end)
fechas_1
```

```python
# si queremos crear unas fechas dentro de un rango: pd.date_range

#me crea un rango DIARIO(freq='D') desde fecha inicio (start) hasta que se cumpla numero de veces de parametro periods
fechas_2 = pd.date_range(start=fecha_modificada,freq='D', periods=8) 
fechas_2
```

```python
#me crea un rango MENSUAL (freq='M') desde fecha inicio (start) hasta que se cumpla numero de veces de parametro periods

fechas_3 = pd.date_range(start= fecha_modificada, periods= 8, freq='M')
fechas_3
```

```python
#una vez tenemos una fecha tipo datetime podemos jugar con ella y cambiarla a otros formatos: strftime

fecha_modificada = pd.to_datetime('03/01/2020',dayfirst=True) #la conversion a tipo datetime

print(fecha_modificada)

mes_inicio = fecha_modificada.strftime('%Y-%m') #aca le decimos que queremos el año - mes solamente
mes_inicio

print(mes_inicio)
```

```python
fechas_4 = pd.period_range(start=mes_inicio, periods=8, freq='M')
fechas_4
```
