Ejercicio: Tenemos una tienda de ropa y queremos contar cuantos productos tenemos que valen menos de 3500 usd y cuantos valen desde 3500 usd en adelante. Previamente clasificamos los precios de la ropa en una lista.
¿Cómo podemos usar las estructuras FOR e IF para ayudarnos a hacer este proceso?


```python
#Respuesta usando for

precios_ropa=[1000,1255,1510,1765,2020,2275,2530,2785,3040,3295,3550,3805,4060,4315,4570,4825,5080,5335,5590,5845,6100,6355,6610,6865]


CONTAR_MENOR=0
CONTAR_MAYOR=0

for i in precios_ropa:
  if i<3500:
    CONTAR_MENOR+=1
  else:
    CONTAR_MAYOR+=1

print("Cantiad de productos con precio mayores a 3500 son " + str(CONTAR_MAYOR))
print("Cantiad de productos con precio menores a 3500  son " + str(CONTAR_MENOR))
```

```python
#Respuesta usando while

precios_ropa=[1000,1255,1510,1765,2020,2275,2530,2785,3040,3295,3550,3805,4060,4315,4570,4825,5080,5335,5590,5845,6100,6355,6610,6865]

indice=0
CONTAR_MENOR=0
CONTAR_MAYOR=0


while indice<len(precios_ropa):
  if precios_ropa[indice]<3500:
    CONTAR_MENOR+=1
  else:
     CONTAR_MAYOR+=1
  indice+=1


print("Cantiad de productos con precio mayores a 3500 son " + str(CONTAR_MAYOR))
print("Cantiad de productos con precio menores a 3500  son " + str(CONTAR_MENOR))
```
