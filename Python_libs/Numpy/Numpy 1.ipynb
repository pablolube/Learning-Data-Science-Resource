###CONCATENAR - FUNCION CONCATENATE

Juntar arreglos, tienen que tener la misma deimensiones

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6],[7,8]])

print("primer array")
print(a)
print("")
print("segundo array")
print(b)
```

```python
np.concatenate((a,b), axis=0) #axis=o significa unir filas de mis arrays
```

```python
np.concatenate((a,b), axis=1) #axis=o significa unir columnas de mis arrays
```

###CONCATENATE - VSTACK / HSTACK

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6],[7,8]])

print("primer array")
print(a)
print("")
print("segundo array")
print(b)
```

```python
Array_apilados_h = np.hstack([a, b])
Array_apilados_h
```

```python
Array_apilados_V = np.vstack([a, b])
Array_apilados_V
```

### SPLIT

Consiste en desarmar o partir los arreglos.
Puede pensarse como la operaciÃ³n inversa a la concatenaciÃ³n.
Las siguientes funciones nos ayudan con esto:

```python
n_array = np.array([1,2,10,20,40,60])

print(n_array)
```

```python
A1,A2,A3=np.split(n_array,[2,4],axis=0)  #aca le digo que lo divida en 3 partes: hasta antes del indice 2, desde el 2 hasta antes del 4 y desde el indice 4

print (A1)
print (A2)
print (A3)
```

```python
narray = np.array([[1,2,10,20,40,60],
                   [100,200,1000,2000,4000,6000],
                   [1,2,10,20,40,60],
                   [100,200,1000,2000,4000,6000]])

print(narray)
print("")


narray1,narray2,narray3= np.split(narray,[2,4],axis=0)  #me crea 3 arrays

print(narray1)
print(" espacio ")
print(narray2)
print(" espacio ")
print(narray3)
```

HSPLIT

```python
n_array = np.array([[1,2,10,20,40,60],
                    [100,200,1000,2000,4000,6000],
                    [150,250,1050,2050,4050,6050]])

print(n_array) #tengo un array que es una matriz de 3 filas y 6 columnas
```

```python
na1,na2,na3=np.hsplit(n_array, 3)  #me quedan 3 arrays cada uno es una matriz de 3 filas y 2 columnas
print(na1)
print(na2)
print(na3)
```

VSPLIT

```python
narray = np.array([[1,2,10,20,40,60],
                   [100,200,1000,2000,4000,6000],
                   [150,250,1050,2050,4050,6050]])

print(narray) #tengo un array que es una matriz de 3 filas y 6 columnas
```

```python
np.vsplit(narray,3) #me quedan 3 arrays cada uno de 1 fila y 6 columnas
```

## AGREGACIONES

UTILIZO UN FOR PARA CALCULAR PROMEDIO

```python
Array_al = np.array([7,4,7,8,3,1,7,4,0,6])

suma = 0
cantidad_elementos=0

for i in Array_al:
  suma += i
  cantidad_elementos+=1

promedio = suma/ cantidad_elementos

print("suma de todos los elementos:",suma)
print("cantidad de elementos",np.size(Array_al))
print("promedio",promedio)
```

UTILIZO AGREGACION PARA CALCULAR PROMEDIO

```python
Array_al.sum() #suma de todos los elementos del array
```

```python
Array_al.mean() #promedio de los elementos del array
```

```python
Array_al.max() #valor mas alto del array
```

```python
Array_al.min() #valor mas alto del array
```
